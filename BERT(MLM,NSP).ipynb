```python
import torch
from transformers import BertTokenizer,BertForMaskedLM,BertForNextSentencePrediction

#creation of tokenizer,model_mlm,model_nsp
tokenizer=BertTokenizer.from_pretrained('bert-base-uncased')
model_mlm=BertForMaskedLM.from_pretrained('bert-base-uncased')
model_nsp=BertForNextSentencePrediction.from_pretrained('bert-base-uncased')
```

# **MLM masked Language modeling**

```python
#Masked LM
def predicted_word(input_text):
  inputs=tokenizer(input_text,return_tensors="pt")
  input_ids=inputs["input_ids"]
  mask_token_index=torch.where(input_ids==tokenizer.mask_token_id)[1]

  with torch.no_grad():
    outputs=model_mlm(**inputs)


  logits=outputs.logits

  mask_token_logit=logits[0,mask_token_index,:]

  predicted_logits=torch.argmax(mask_token_logit,dim=-1)

  prediction=tokenizer.decode(predicted_logits)
  return prediction
```

```python
text="i always want to go to [MASK] and play."
X=predicted_word(text)
print(X)
print("text input before prediction :",text)
text=text.replace('[MASK]',X)
print("text after before prediction :",text)
```

# **NSP (Next Sentence Prediction)**

```python
#nsp

def nsp(sentence_1,sentence_2):
  inputs=tokenizer.encode_plus(sentence_1,sentence_2,return_tensors="pt")
  input_ids=inputs["input_ids"]
  token_type_ids=inputs["token_type_ids"]
  attention_mask=inputs["attention_mask"]

  with torch.no_grad():
    outputs=model_nsp(input_ids=input_ids,token_type_ids=token_type_ids,attention_mask=attention_mask)
  logits=outputs.logits
  predicted=torch.argmax(logits,dim=-1).item()
  return predicted
```

```python
sent_1="hello my name is Taran"
sent_2="killing"

y=nsp(sent_1,sent_2)
print(y)
```
